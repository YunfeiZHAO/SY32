## daisy
Calcul des caractristiques DAISY et redimensionne en tableau 2D : k x n.

Une image est reprsente par k vecteurs de dimension n.

La dimension n du descripteur vaut (rings*histograms+1)*orientations=200

```python
def d_daisy(im):
    descr = daisy(im, step=16, radius=32, rings=3, histograms=8,
                  orientations=8, visualize=False)
    return descr.reshape((descr.shape[0]*descr.shape[1], descr.shape[2]))
```

Tableau contenant les descripteurs DAISY pour l'ensemble des 300 images

```python
descr = np.empty((0, 200))
```

Tableau d'indices permettant de retrouver les descripteurs associs  chaqueimage. 

Les descripteurs de l'image i sont compris entre descr_idx[i-1] et descr_idx[i].

```python
descr_idx = np.zeros(300)
```

## Calcul des caractristiques pour toutes les images

```python
for i in range(300):
    im = rgb2gray(imread(f'images/{i:03d}.jpg'))
    descr = np.concatenate((descr, d_daisy(im)))
    descr_idx[i] = len(descr)
```

## Clustering pour construire le dictionnaire de mots visuels

```python
n_clust = 20
kmeans = KMeans(n_clusters=n_clust)
kmeans.fit(descr)
```

## Calcul d'un histogramme  partir d'une liste d'indices de mots visuels

```python
# numpy bincount: returns the size of each 'bin' for each value from 0 to the largest value in the array it is called on.
def mots2hist(mots, n_clust):
    hist = np.bincount(mots, minlength=n_clust)
    hist = hist/np.sum(hist)
    return hist.reshape((1, len(hist)))
```

## sous la forme d'un histogramme sur le dictionnaire
La fonction predict retourne l'indice du mot visuel associ au descripteur


```python
mots = kmeans.predict(descr)
```

```python
descr_hist = np.zeros((300, n_clust))
```

```python
for i in range(300):
    i_start = int(descr_idx[i-1]) if i > 0 else 0
    i_end = int(descr_idx[i])
    descr_hist[i, :] = mots2hist(mots[i_start:i_end], n_clust)
```

## Recherche des images les plus proches

```python
img = ['crab.jpg', 'kangaroo.jpg', 'cougar.jpg']
```

```python
for i in range(len(img)):
    im0 = rgb2gray(imread(img[i]))
    descr0 = d_daisy(im0)
    mots0 = kmeans.predict(descr0)
    descr_hist0 = mots2hist(mots0, n_clust)
    
    d = distance_matrix(descr_hist0, descr_hist)
    d_i = np.argsort(d[0,:])
    plt.figure(img[i])
    plt.clf()
    for j in range(9):
        plt.subplot(3, 3, j+1)
        plt.imshow(imread(f'images/{d_i[j]:03d}.jpg'))
```
